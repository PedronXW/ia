## Transformando dados

```python
## Transforma o campo 'TotalCharges' para numérico, tratando erros(Dados que não possam ser convertidos) como NaN
df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'], errors='coerce')
```

```python
df_contracts.info()
```

## Renomeando colunas

```python
## Renomeando colunas, somente retorna a coluna renomeada, mas não altera o dataframe original
df_customers.rename(columns={'SeniorCitizen': 'Above65yo'})
```

```python
## Renomeando colunas e alterando o dataframe original
df_customers.rename(columns={'SeniorCitizen': 'Above65yo'}, inplace=True)
```

```python
## Renomeando colunas com lista para multiplas colunas
df_customers.columns = ['IDCliente', 'Genero', 'Mais65Anos', 'TemParceiro', 'TemDependentes']
```

```python
df_customers.info()
```

## Unificar dataframes

```python
##Precisamos primeiro garantir que a coluna de identificação do cliente que estamos usando como referência é a mesma em todos os dataframes(customerId renomeado para IDCliente)
df_services.rename(columns={'customerID': 'IDCliente'}, inplace=True)
df_contracts.rename(columns={'customerID': 'IDCliente'}, inplace=True)
```

```python
## Unificando os tres dataframes em um só, usando a coluna IDCliente como referência
df_churn = df_customers.merge(df_services, on='IDCliente').merge(df_contracts, left_on='IDCliente', right_on='IDCliente')
```

```python
df_churn.info()
```

## Detecção de valores ausentes

```python
## Detecção de valores ausentes, verificando se existem colunas com valores ausentes
df_churn.isna().sum()
```

```python
## Quantas linhas tem pelo menos um valor ausente
df_churn[df_churn.isna().any(axis=1)]
```

```python
## Quantas colunas tem pelo menos um valor ausente
df_churn.isna().any(axis=0).sum()
```

## Remover valores ausentes

```python
## Remove colunas
df_churn.drop(['TotalCharges'], axis=1, inplace=True)
```

```python
## Remover colunas com valores ausentes
df_churn.dropna(axis=0, inplace=True)
```

```python
## Remover coluna com todos os valores ausentes
df_churn.dropna(axis=1, how='all', inplace=True)
```

```python
## Remover linhas com pelo menos um valor ausente
df_churn.dropna(axis=0, inplace=True)
```

```python
## Remover linhas com todos os valores ausentes
df_churn.dropna(axis=0, how='all', inplace=True)
```

## Inputando valores ausentes

```python
## Preenchimento de valores ausentes
df_churn.fillna(0)
```

```python
## Preenchimento de valores ausentes conforme a coluna
df_churn.fillna({'Genero': 'Não declarado', 'TotalCharges': 0}, inplace=True)
```

```python
## Preenchimento de valores ausentes com a média da coluna
df_churn.fillna(df_churn.mean(), inplace=True)
## ou
df_churn.fillna(value={'TotalCharges': df_churn['TotalCharges'].mean()}, inplace=True)
```
